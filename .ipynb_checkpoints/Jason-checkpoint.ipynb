#### Imputation

```python
# impute
first_na = df[df['sum_trade_1s'].isna()].index[0]
print(first_na)
df[['ask_qty','sum_trade_1s', 'last_trade_time']].iloc[first_na-40:first_na+1]
```

Trade of qty 1 executed at time after row 18 because last_trade_time is no longer increasing

At row 53, the previous 112 lots is no longer counted in this second therefore sum_trade_1s = 1

At row 56, it has been >1s since the latest trade of qty 1, therefore there are 0 trades by this entry

##### Conclude: Fill NaNs with 0, assuming standard practice of leaving 0 as blank when data  for sum_trade_1s is being collected

```python
df['sum_trade_1s'] = df['sum_trade_1s'].fillna(0)
df[(df['last_trade_time'].isna()) & (df['sum_trade_1s']!=0)]
```

when last_trade_time is NaN, sum_trade_1s is always 0

```python
# explore NaNs
sample_na = df[df['last_trade_time'].isna()].index[1]
print("sample id of row with NaN:",sample_na)
df[['timestamp','sum_trade_1s', 'last_trade_time']].iloc[sample_na-5:sample_na+60]
```

NaN goes away when trade is executed (tested for 3 sample_na)
##### Conclude: Fill NaNs with last known + diff in timestamp

```python
# fill nans
na_ids = list(df[df['last_trade_time'].isna()].index)
for na_id in na_ids:
    timestamp_diff = df.iloc[na_id]['timestamp'] - df.iloc[na_id-1]['timestamp']
    non_na_last_trade_time = df.iloc[na_id-1]['last_trade_time']
    df.iloc[na_id, df.columns.get_loc('last_trade_time')] = non_na_last_trade_time + timestamp_diff.seconds+timestamp_diff.microseconds/1000000
```

```python
# plot against time function
def plot_against_time(features):
    fig, ax = plt.subplots(figsize=(20,5))
    for feature in features:
        ax.plot_date(df.timestamp, df[feature], alpha=0.01)
    legend = ax.legend()
    for lh in legend.legendHandles:
        lh._legmarker.set_alpha(1)

# trade/ask
plot_against_time(['trade_price','bid_price', 'ask_price'])
```

bid & ask follows same trend

trade_price has significant grouped outliers as well as scattered ones

at start of 03:00, last trade price suddenly drops 10 ticks even though last trade price, bid price and ask price have been increasing

#### Handling outliers

```python
df[(df['bid_price'] - df['trade_price'] > 50) & (df['trade_price'] == 23300)]
df.iloc[59803-10:59803+10]
```

##### Conclude: bid/ask qty and prices do not seem to be a factor of anomalous (last) trade_prices, exclude for initial run of model

```python
# remove grouped outliers
old_len = len(df)
df1 = df[df['timestamp'] < datetime.datetime(2018,2,1,3,0,15,483773)]
df2 = df[(df['timestamp'] > datetime.datetime(2018,2,1,3,15,22,797721)) & (df['timestamp'] < datetime.datetime(2018,2,1,4,30,5,998897))]
df3 = df[df['timestamp'] > datetime.datetime(2018,2,1,4,45,20,128272)]
df = pd.concat([df1,df2,df3])
print(len(df)-old_len, "grouped outliers removed")
```

```python
# remove scattered outliers
print(len(df) - len(df[np.abs(df.trade_price-df.trade_price.mean()) <= (2*df.trade_price.std())]), "scattered outliers removed")
df = df[np.abs(df.trade_price-df.trade_price.mean()) <= (2*df.trade_price.std())]
print("df reduced to",len(df))
```

## Feature Creation

```python
# bid/ask spread as percentage - measure of liquidity
df['spread'] = (df['ask_price'] - df['bid_price']) / df['ask_price']
plot_against_time(['spread'])

# delta - measure of aggressiveness, if positive means trading more on the ask, more risky
df['delta'] = df['ask_qty'] - df['bid_qty']
plot_against_time(['delta'])
```

##### one-hot encode day, hour (few features)

```python
# day
df['day_is_31'] = df.timestamp.dt.day
df.day_is_31.replace([31,1], [1,0], inplace=True)
# hour
df = pd.concat([df, pd.get_dummies(df.timestamp.dt.hour, prefix='hour')], axis=1)
```

##### circular encode min, sec (many features, continuous values from 59 to 0)

```python
# min
df['minute_sin'] = np.sin(2 * np.pi * df.timestamp.dt.minute/59.0)
df['minute_cos'] = np.cos(2 * np.pi * df.timestamp.dt.minute/59.0)
# sec
df['second_sin'] = np.sin(2 * np.pi * df.timestamp.dt.second/59.0)
df['second_cos'] = np.cos(2 * np.pi * df.timestamp.dt.second/59.0)
ax = df.plot.scatter('second_sin', 'second_cos').set_aspect('equal')
```

```python
# microsecond
df['microseconds'] = df.timestamp.dt.microsecond
```

```python
# trade occurred
df['trade_occurred'] = (df.last_trade_time < df.last_trade_time.shift()).astype(int)
# ask price increased
df['ask_increased'] = (df.ask_price > df.ask_price.shift()).astype(int)
# bid price decreased
df['bid_decreased'] = (df.bid_price < df.bid_price.shift()).astype(int)
# price changed
df['price_changed'] = ((df.ask_price != df.ask_price.shift()) | (df.bid_price != df.bid_price.shift())).astype(int)
```

## Multi-label, multi-class classification

#### AdaBoost

```python
# AdaBoost
kf = StratifiedShuffleSplit(n_splits=3)

list_score = []
list_score_per_class= []

for train, test in kf.split(X, y):
    X_train, X_test, y_train, y_test = X.iloc[train], X.iloc[test], y.iloc[train], y.iloc[test]

    clf = AdaBoostClassifier()
    multi_target_classifier = MultiOutputClassifier(clf, n_jobs=-1)
    multi_target_classifier.fit(X_train, y_train)
    preds = multi_target_classifier.predict(X_test)

    f1s = []
    for i in range(3):
        f1 = f1_score(np.array(y_test)[:,i], preds[:,i], average='weighted')
        f1s.append(f1)
    print("F1 Score:")
    print(f1s)
    list_score_per_class.append(f1s)

list_score_per_class = np.array(list_score_per_class)
print('Mean score per round: %s' % list_score_per_class.mean(axis=0))
```

#### Random Forest

```python
# random forest
kf = StratifiedShuffleSplit(n_splits=3)

list_score = []
list_score_per_class= []

for train, test in kf.split(X, y):
    X_train, X_test, y_train, y_test = X.iloc[train], X.iloc[test], y.iloc[train], y.iloc[test]

    clf = RandomForestClassifier(n_estimators = 10, random_state = 42)
    multi_target_classifier = MultiOutputClassifier(clf, n_jobs=-1)
    multi_target_classifier.fit(X_train, y_train)
    preds = multi_target_classifier.predict(X_test)

    f1s = []
    for i in range(3):
        f1 = f1_score(np.array(y_test)[:,i], preds[:,i], average='weighted')
        f1s.append(f1)
    print("F1 Score:")
    print(f1s)
    list_score_per_class.append(f1s)

list_score_per_class = np.array(list_score_per_class)
print('Mean score per round: %s' % list_score_per_class.mean(axis=0))
```

Mean F1-score across 3 rounds of stratified shuffle split cross validation for labels 

1s_side = 0.95517089, 3s_side = 0.95638218, 5s_side = 0.95632912
